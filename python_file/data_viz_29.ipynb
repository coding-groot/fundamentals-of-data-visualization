```python
from nycflights13 import flights
```

```python
delay = flights.groupby('carrier')['arr_delay'].agg(**{'mean_delay':'mean'}).reset_index()
```

```python
targets = ["OO", "HA", "YV", "F9", "AS", "FL", "VX"]
delay = delay[-delay['carrier'].isin(targets)]
delay = delay.sort_values(by='mean_delay')
```

```python
delay
```

```python
clrs = ['red' if (x < 2) else 'grey' for x in delay['mean_delay']]
sns.barplot(x="mean_delay", y="carrier", data=delay,palette=clrs)
plt.show()
```

```python
import altair as alt
import pandas as pd

source = pd.DataFrame([
      {'animal': 'dog'},{'animal': 'dog'},{'animal': 'dog'},
      {'animal': 'dog'},{'animal': 'dog'},{'animal': 'dog'},
      {'animal': 'dog'},{'animal': 'dog'},{'animal': 'dog'},
      {'animal': 'cat'},{'animal': 'cat'},{'animal': 'cat'},
      {'animal': 'cat'},{'animal': 'cat'},{'animal': 'cat'},
      {'animal': 'cat'},{'animal': 'cat'},
      {'animal': 'fish'},{'animal': 'fish'},
      {'animal': 'bird'}])

alt.Chart(source).mark_text(size=45, baseline='middle').encode(
    alt.X('x:O', axis=None),
    alt.Y('animal:O', axis=None),
    alt.Text('emoji:N')
).transform_calculate(
    emoji="{'dog': '🐶', 'cat': '🐱', 'fish': '🐟','bird':'🐤'}[datum.animal]"
).transform_window(
    x='rank()',
    groupby=['country', 'animal']
).properties(width=550, height=140)
```
