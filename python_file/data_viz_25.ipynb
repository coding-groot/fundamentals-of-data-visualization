```python
iris = sns.load_dataset("iris")
iris
```

image.png

```python
fig = plt.figure(figsize=(6, 5))
p = sns.kdeplot(data=iris,x='sepal_length',hue='species',palette='gray')

lss = [':', '--', '-.']

handles = p.legend_.legendHandles[::-1]

for line, ls, handle in zip(p.lines, lss, handles):
    line.set_linestyle(ls)
    handle.set_ls(ls)
```

```python
sns.kdeplot(data=iris,x='sepal_length',hue='species')
plt.show()
```

```python
sns.kdeplot(data=iris,x='sepal_length',hue='species',fill=True)
plt.show()
```
