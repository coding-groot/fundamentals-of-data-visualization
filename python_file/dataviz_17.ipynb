```python
raw_data ={'city name':['Micronesia','Tonga','Vanuatu','Samoa','Solomon Islands','Fiji','Papua New Guinea','New Zealand','Australia'],
           'gdp':[329,437,837,844,1273,5054,21810,200800,1390000]}
gdp = DataFrame(raw_data)
```

```python
import plotly.express as px
fig = px.bar(gdp,x='gdp',y='city name',orientation='h')
fig.show()
```

```python
import plotly.express as px
fig = px.bar(gdp,y='city name', x='gdp',log_x = True)
fig.show()
```

```python
import plotly.graph_objects as go
fig= go.Figure()
fig = px.scatter(gdp, x="gdp", y="city name", text="city name", log_x=True, size_max=60)
fig.update_traces(textposition='top center')
fig.update_traces(marker_size=15)
fig.update_yaxes(showticklabels=False)
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
gdp['ratio'] = np.log2(gdp['gdp']/21810)
x = sns.barplot(data =gdp, x='ratio', y='city name',palette=sns.color_palette("Set3",1))
x.set(xticklabels=[],yticklabels=[],xlabel=None,ylabel=None)
```

```python
raw_data ={'county':['Providence','Kent','Washington','Newport','Bristol'],
           'population':[627000,166000,127000,82900,49900]}
population = DataFrame(raw_data)
```

```python
fig = px.pie(population, values='population', names='county')
fig.update_traces(textinfo='value')
fig.show()
```

```python
fig = px.bar(population,y='county', x='population')
fig.show()
```

```python
fig = px.treemap(population,path = ['county'],values='population')
#fig.add_traces(textinfo='value')
fig.show()
```
