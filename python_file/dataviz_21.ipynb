```python
sns.lineplot(x=flights['year'],y=flights['passengers'],ci=None)
plt.show()
```

```python
flight = flights[(flights['year']==1950)|(flights['year']==1960)]
p1 = list(flight['passengers'][:12]/sum(flight['passengers'][:12]))
p2 = list(flight['passengers'][12:]/sum(flight['passengers'][12:]))
flight['pro'] = p1 + p2
sns.pointplot(x="year", y="pro", hue = 'month',data=flight)
```
